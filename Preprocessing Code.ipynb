# New Section

```python
import pandas as pd
```

```python
daily_df = pd.read_csv("/content/טמפרטורה, התאדות, זהירת שמש - יומי !.csv")
```

```python
daily_df
```

```python
daily_df.columns = ['station', 'date', 'max temp', 'min temp', 'near ground temp', 'Cumulative evaporation 12 hours (mm)', 'Daily evaporation from A tub (mm)', 'Duration of sun exposure (minutes)']
```

```python
daily_df
```

```python

# Step 1: Convert "date" column to pandas datetime
daily_df['date'] = pd.to_datetime(daily_df['date'], format='%d/%m/%Y')
daily_df
```

```python
daily_df['station'] = daily_df['station'].str.replace(r'\d+', '')
daily_df['new station'] = daily_df['station'].str.extractall(r'\b(\w{3,})\b').groupby(level=0).agg(' '.join)
```

```python
daily_df['new station'] = daily_df['new station'].str.replace('מאוישת', '')
daily_df['new station'] = daily_df['new station'].str.replace('ישנה', '')
```

```python
daily_df['new station'] = daily_df['new station'].str.replace('חווה', '')
```

```python
daily_df['new station'] = daily_df['new station'].str.strip()
```

```python
daily_df['new station'].unique()
```

```python
# Step 2: Extract month from the date and create a new column
daily_df['month'] = daily_df['date'].dt.month
daily_df
```

```python
# Step 3: Extract year from the date and create a new column
daily_df['year'] = daily_df['date'].dt.year
```

```python
station_avg_temp = daily_df.groupby('new station').mean()
```

```python
fill_cols = ['max temp', 'min temp', 'near ground temp', 'Cumulative evaporation 12 hours (mm)', 'Daily evaporation from A tub (mm)', 'Duration of sun exposure (minutes)']
for col in fill_cols:
  daily_df[col].fillna(daily_df['new station'].map(station_avg_temp[col]), inplace=True)
```

```python
# Step 4: Group by "station" and "month", then calculate the mean for each group
average_by_station_month = daily_df.groupby(['new station', 'month', 'year']).mean().reset_index()
average_by_station_month
```

```python
average_by_station_month.to_csv("average_by_station_month_year.csv")
```
